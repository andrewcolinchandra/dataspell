```python
from sqlalchemy import create_engine
import geopandas
db_connection_url = "postgresql://postgres@localhost:5432/acc"
con = create_engine(db_connection_url)
sql = 'SELECT * FROM public.checkindata'
checkindata = geopandas.read_postgis(sql=sql, con=con, geom_col='CheckInCoordinates', crs='epsg:4326')
```

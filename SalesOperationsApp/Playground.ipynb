First import all the necessary libraries

```python
import pandas as pd
```

Set up the connection to Firestore

```python

import firebase_admin
from firebase_admin import credentials

cred = credentials.Certificate("salesapps-cc685-firebase-adminsdk-dgpq2-77f3b0918e.json")
firebase_admin.initialize_app(cred)

from firebase_admin import firestore
db=firestore.client()

```

```python
cid = list(db.collection(u'check_in_dev').stream())
cid_dict = list(map(lambda x: x.to_dict(), cid))
df = pd.DataFrame(cid_dict)
df
```

```python
df1 = pd.json_normalize(cid_dict)
df1
```

```python
df2 = pd.json_normalize(df1['competitor.data'])
df2
```

```python
df3 = pd.json_normalize(df2[0])
df3
```

```python
tf1 = pd.json_normalize(cid_dict, record_path='competitor.data')
tf1
```
